```python
import  pandas  as  pd 
import  json
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

## 3.1 Исследование целевой переменной

```python
df = pd.read_csv('articles.csv')
```

```python
df
```

```python
df.drop(['Unnamed: 0' , 'description' , 'rating' , 'industries'] ,axis = 1, inplace = True)
```

```python
df['company_name'] = df['company_name'].apply(lambda x: x[13:])
```

```python
df['company_name'].value_counts()
```

```python
df
```

```python
import json

with open('Target.json', 'r', encoding='utf-8') as f:
    target = json.load(f)
```

```python
df_json = pd.DataFrame(target['data'])
```

```python
dict1 = {}
for company, nomination in zip(df_json['Сompany'], df_json['Nominations']):
    dict1[company] = nomination
```

```python
df_json
```

```python
from fuzzywuzzy import fuzz
```

```python
matches = []
for company_name in df['company_name']:
    max_ratio = 0
    max_company_name = None
    for json_company_name in df_json['Сompany']:
        ratio = fuzz.ratio(company_name, json_company_name)
        if ratio > max_ratio:
            max_ratio = ratio
            max_company_name = json_company_name
    matches.append(max_company_name)

# Add the matches to the DataFrame
df['target'] = matches
```

```python
df['target'] = df['target'].apply(lambda x: dict1[x])
```

```python
import  numpy  as  np
```

```python
emb  = np.load('emb.npy')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['target'] = le.fit_transform(df['target'])
```

```python
df
```

```python
labels = df["target"].values
train_features, test_features, train_labels, test_labels = train_test_split(emb, labels , random_state = 42 , test_size = 0.2)
```

```python
df["target"].value_counts()
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Предположим, что train_labels - это массив меток для обучения
classes = np.unique(train_labels)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=train_labels)

# Используем другое имя для словаря весов, чтобы избежать конфликта типов
class_weight_dict = dict(zip(classes, weights))

# Теперь class_weight_dict - это словарь весов для классов
print(class_weight_dict)
```

```python
lr_clf = LogisticRegression(class_weight = class_weight_dict)
lr_clf.fit(train_features,  train_labels)
```

```python
lr_clf.score(test_features, test_labels)
```

```python
from catboost import CatBoostClassifier
cat_cls = CatBoostClassifier( class_weights = class_weight_dict , iterations=2000 , early_stopping_rounds = 400 ,  random_state=42 )
cat_cls.fit(train_features,  train_labels)
```

```python
from sklearn.metrics import classification_report
```

```python
cat_cls.score(test_features, test_labels)
```

```python
y_pred = cat_cls.predict(test_features)
```

```python

print(classification_report(test_labels, y_pred))
```
